```python
def similar_product(id, n):
  sim_scores = list(enumerate(euclidean_matrix[id]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=False)
  sim_scores = sim_scores[1:(n+1)]
  product_indices = [i[0] for i in sim_scores]
  score = [i[1] for i in sim_scores]
  a = pd.DataFrame(df['title'].iloc[product_indices])
  a['Euclidean Score'] = score
  a['main_category'] = pd.DataFrame(df['main_category'].iloc[product_indices])
  a['subcategory'] = pd.DataFrame(df['subcategory'].iloc[product_indices])
  return a
```

```python
def precision(id, n):
  sim_scores = list(enumerate(euclidean_matrix[id]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=False)
  sim_scores = sim_scores[1:(n+1)]
  product_indices = [i[0] for i in sim_scores]
  score = [i[1] for i in sim_scores]
  a = pd.DataFrame(df['title'].iloc[product_indices])
  a['subcategory'] = pd.DataFrame(df['subcategory'].iloc[product_indices])
  true = df['subcategory'].iloc[id]
  precision = len(a[a['subcategory'] == true])
  return precision/n

def averagePrecision(id):
  map = precision(id, 5)
  map += precision(id, 10)
  map += precision(id, 20)
  map = map/3
  return map

def meanAveragePrecision():
  total_precision = 0
  for i in range(len(index)):
     total_precision += averagePrecision(index[i])

  total_precision = total_precision/20
  return total_precision
```
