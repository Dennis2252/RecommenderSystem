```python
def tokenize(text):
    tokens = word_tokenize(text.lower())
    return tokens
```

```python
def stem(text):
    tokens = word_tokenize(text.lower())
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return " ".join(stemmed_tokens)
```

```python
def lemmatizing(text):
    tokens = word_tokenize(text.lower())
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in stemmed_tokens]
    return " ".join(lemmatized_tokens)
```

```python
def preprocess_keyword(text):
    tokens = word_tokenize(text.lower())
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in stemmed_tokens]
    return " ".join(lemmatized_tokens)
```

```python
def similar_product(id, n):
  sim_scores = list(enumerate(cosine_similarities[id]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:(n+1)]
  product_indices = [i[0] for i in sim_scores]
  score = [i[1] for i in sim_scores]
  a = pd.DataFrame(df['title'].iloc[product_indices])
  a['Cosine Similarity Score'] = score
  a['main_category'] = pd.DataFrame(df['main_category'].iloc[product_indices])
  a['subcategory'] = pd.DataFrame(df['subcategory'].iloc[product_indices])
  return a
```

```python
def precision(id, n):
  sim_scores = list(enumerate(cosine_similarities[id]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:(n+1)]
  product_indices = [i[0] for i in sim_scores]
  score = [i[1] for i in sim_scores]
  a = pd.DataFrame(df['title'].iloc[product_indices])
  a['subcategory'] = pd.DataFrame(df['subcategory'].iloc[product_indices])
  true = df['subcategory'].iloc[id]
  precision = len(a[a['subcategory'] == true])
  return precision/n
```

```python
precision(5, 5)
```

```python
def averagePrecision(id):
  map = precision(id, 5)
  map += precision(id, 10)
  map += precision(id, 20)
  map = map/3
  return map
```

```python
import math

def recall(id, n):
  sim_scores = list(enumerate(cosine_similarities[id]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  relevant = sim_scores[1:]
  sim_scores = sim_scores[1:(n+1)]

  product_indices = [i[0] for i in sim_scores]
  relevant_indices = [i[0] for i in relevant]

  score = [i[1] for i in sim_scores]
  score_relevant = [i[1] for i in relevant]

  a = pd.DataFrame(df['title'].iloc[product_indices])
  a['subcategory'] = pd.DataFrame(df['subcategory'].iloc[product_indices])
  a['Cosine Similarity'] = score

  b = pd.DataFrame(df['title'].iloc[relevant_indices])
  b['subcategory'] = pd.DataFrame(df['subcategory'].iloc[relevant_indices])
  b['Cosine Similarity'] = score_relevant

  true = df['subcategory'].iloc[id]
  precision = len(a[a['subcategory'] == true])

  if len(b[b['subcategory'] == true]) == 0:
    return 0
  else:
    recall = precision/(len(b[b['subcategory'] == true]))
    # print(len(b[b['subcategory'] == true]) - precision)
    # print(len(a[a['subcategory'] == true]))
    # print(threshold)
    return recall
```
